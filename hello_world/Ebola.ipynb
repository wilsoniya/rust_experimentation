```python
df = pd.read_csv('SEIRDh_data_flat.txt', sep='\t')
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

col = 'tot_case'
areas = df.groupby('area')[col].sum().copy().reset_index()
areas = areas[areas[col].notnull()]
areas = areas.sort(col)
areas = areas.set_index('area')
areas.plot(kind='bar')
 
col = 'tot_death'
areas = df.groupby('area')[col].sum().copy().reset_index()
areas = areas[areas[col].notnull()]
areas = areas.sort(col)
areas = areas.set_index('area')
areas.plot(kind='bar')
```

```python
col = 'tot_case'
monthly = df.groupby('date')[col].sum().copy()
monthly.sort_index()
monthly.plot(logy=True)

col = 'tot_death'
monthly = df.groupby('date')[col].sum().copy()
monthly.sort_index()
monthly.plot(logy=True)
```
